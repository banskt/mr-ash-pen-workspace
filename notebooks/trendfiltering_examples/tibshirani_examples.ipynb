```python
# Hills example
n = 128
p = 128
order = 3
std = 0.001
#bidx = np.array([30, 50, 80, 90, 100, 110, 120])
bidx = np.array([5, 20, 40])
bfix = np.array([3, -27, 19683])
s = bidx.shape[0]
X, y, Xtest, ytest, btrue, snr = model.changepoint(n, p, s, std, seed = 100,
                                                  order = order, signal = "fixed",
                                                  bfix = bfix,
                                                  bidx = bidx)
print(snr)

fig = plt.figure()
ax1 = fig.add_subplot(111)

xvals = np.arange(n)
ypred = np.dot(Xtest, btrue)
ax1.scatter(xvals, y, edgecolor = 'black', facecolor = 'white')
ax1.plot(xvals, ypred, color = 'dodgerblue')
plt.show()
```

```python
G = np.zeros((n, n))
for i in range(n):
    for j in range(order + 1):
        G[i, j] = np.power(i + 1, j) / np.power(n, j)
    for j in range(order + 1, n):
        if i <= (j - (order + 1) / 2):
            print("Zero", i+1, j+1)
            G[i, j] = 0
        else:
            print("Nonzero", i+1, j+1)
            khalf = (order + 1) / 2
            G[i, j] = np.power((i - j + khalf)/n, order)
```

```python
def plot_trendfilter(X, y, beta, ytest, bhat,
                              intercept = 0, title = None):
    n = y.shape[0]
    p = X.shape[1]

    ypred = np.dot(X, bhat) + intercept
    fig = plt.figure(figsize = (12, 6))
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    ax1.scatter(np.arange(n), ytest, edgecolor = 'black', facecolor='white')
    ax1.plot(np.arange(n), ypred)
    ax1.set_xlabel("Sample index")
    ax1.set_ylabel("Y")

    ax2.scatter(np.arange(p), beta, edgecolor = 'black', facecolor = 'white')
    ax2.scatter(np.arange(p), bhat, s = 40, color = 'firebrick')
    ax2.set_xlabel("Sample index")
    ax2.set_ylabel("b")
    
    if title is not None:
        fig.suptitle(title)

    plt.tight_layout()
    plt.show()
```
