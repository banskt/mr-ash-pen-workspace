```python
import numpy as np
import scipy as sp
import networkx as nx
import matplotlib.pyplot as plt
from pygsp import graphs
import pygsp

from pymir import mpl_stylesheet
mpl_stylesheet.banskt_presentation(splinecolor = 'black', dpi = 120)
```

```python
np.random.seed(1000)

W = np.random.uniform(size=(30, 30))  # Full graph.
W[W < 0.93] = 0  # Sparse graph.
W = W + W.T  # Symmetric graph.
np.fill_diagonal(W, 0)  # No self-loops.

G = graphs.Graph(W)
Gnx = nx.from_numpy_matrix(W)

print(f'{G.N} nodes, {G.Ne} edges')
```

```python
fig = plt.figure(figsize = (6, 6))
nx.draw(Gnx)
plt.show()
```

```python
G.set_coordinates(kind = 'spring')
pygsp.plotting.plot_graph(G)
```
